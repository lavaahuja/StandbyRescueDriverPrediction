```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("sickness_table.csv")
data.head()
```

```python
data.describe(include='all')
```

```python

data['date'] =pd.to_datetime(data['date'])
data.info()
```

```python
plt.plot(data['date'], data['calls'], linestyle='-')
plt.show()
```

```python
filtered_df = data[(data['date'].dt.month == 4)& (data['date'].dt.year == 2018)]
filtered_df
```

```python
# Filter the data for April 2018
filtered_df = data[(data['date'].dt.month == 4) & (data['date'].dt.year == 2018)]

# Plot the filtered data
plt.plot(filtered_df['date'], filtered_df['n_sick'], linestyle='-')
plt.show()
```

```python
may_data = data[data['date'].dt.month == 4]
may_data
```

```python
may_data = data[data['date'].dt.month == 5]
years = may_data['date'].dt.year.unique()
no_of_plots = len(years)
fig, axes = plt.subplots(no_of_plots, 1, figsize=(8, 4 * no_of_plots), sharex=False)


for i, year in enumerate(years):
    data_for_year = may_data[may_data['date'].dt.year == year]
    print(data_for_year)
    axes[i].plot(data_for_year['date'], data_for_year['n_sick'],marker='o', linestyle='-', label=f'Year {year}')
    axes[i].set_ylabel('Value')
    axes[i].legend()
plt.xlabel('Date')
plt.suptitle('Line Plots for May Data (Year-wise)')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
data2k16 = data[data['date'].dt.year == 2016]
months = data2k16['date'].dt.month.unique()
months
```

```python
data2k16 = data[data['date'].dt.year == 2016]
months = data2k16['date'].dt.month.unique()
no_of_plots = len(months)
fig, axes = plt.subplots(no_of_plots, 1, figsize=(8, 4 * no_of_plots), sharex=False)


for i, month in enumerate(months):
    data_for_month = data2k16[data2k16['date'].dt.month == month]
    print(data_for_month)
    axes[i].plot(data_for_month['date'], data_for_month['n_sick'],marker='o', linestyle='-', label=f'month {month}')
    axes[i].set_ylabel('Value')
    axes[i].legend()
plt.xlabel('Date')
plt.suptitle('Line Plots for Yearly Data (Month-wise)')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

```

```python
average_sick_calls_2016 = data[data['date'].dt.year == 2016]
months = average_sick_calls_2016['date'].dt.month.unique()
months
```

```python
average_sick_calls_2016 = data[data['date'].dt.year == 2016]
months = average_sick_calls_2016['date'].dt.month.unique()
mean = pd.Series()
for i, month in enumerate(months):
    monthly_data = average_sick_calls_2016[average_sick_calls_2016['date'].dt.month == month]
    mean_x = monthly_data['n_sick'].mean()
    mean = mean.append(pd.Series(mean_x))
plt.bar(months, mean)
plt.show()
```
